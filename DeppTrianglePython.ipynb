```python
def df_to_model_input(df):
    paid = df.pivot_table(index=['lob','group_code','accident_year'], columns='development_lag', values=['incremental_paid'], aggfunc=np.sum)
    reserves = df.pivot_table(index=['group_code','lob','accident_year'], columns='development_lag', values=['case_reserves'], aggfunc=np.sum)

    #triangles = np.append(paid.to_numpy()[:,np.newaxis],reserves.to_numpy()[:,np.newaxis], axis=2)

    triangles=np.stack([paid.to_numpy(),reserves.to_numpy()])

    print(triangles.shape)

    company_code = paid.index.get_level_values(0)

    return [triangles, company_code]
```

```python
from tensorflow.keras.layers import TimeDistributed, GRU, Dense, RepeatVector, LSTM
from tensorflow.keras.models import Sequential

import pandas as pd
import numpy as np
import tensorflow as tf
tf.compat.v1.disable_eager_execution()

from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
import tensorflow.keras.backend as tfk
```

```python

tfk.clear_session()

ay_seq_input  = layers.Input(shape = (9,2), name = 'ay_seq_input' )
company_code_input = layers.Input(shape = 1, name = "company_input")
company_code_embedding = layers.Embedding(200,49)(company_code_input)
company_code_embedding = layers.Flatten()(company_code_embedding)
company_code_embedding = layers.RepeatVector(9)(company_code_embedding)
   

encoded = layers.Masking(mask_value = -99)(ay_seq_input)
encoded = layers.GRU(128, dropout = 0.2, recurrent_dropout = 0.2)(encoded)

concat_layer = lambda x: layers.Concatenate()([x, company_code_embedding])
decoded = layers.RepeatVector(9)(encoded)
decoded = layers.GRU(128, return_sequences = True, dropout = 0.2, recurrent_dropout = 0.2)(decoded)
decoded = layers.Lambda(concat_layer)(decoded)

case_reserves_output = layers.TimeDistributed(layers.Dense(units = 64, activation = "relu"))(decoded)
case_reserves_output = layers.TimeDistributed(layers.Dropout(rate = 0.2))(case_reserves_output)
case_reserves_output = layers.TimeDistributed(layers.Dense(units = 1, activation = "relu"), name = "case_reserves_output")(case_reserves_output)

paid_output = layers.TimeDistributed(layers.Dense(units = 64, activation = "relu"))(decoded)
paid_output = layers.TimeDistributed(layers.Dropout(rate = 0.2))(paid_output)
paid_output = layers.TimeDistributed(layers.Dense(units = 1, activation = "relu"), name = "paid_output")(paid_output)



model = keras.Model(
inputs = [ay_seq_input, company_code_input],
outputs = [paid_output, case_reserves_output],
    name = "DeepTriangle"
)

model.summary()
```

```python
#%%
''' 9. Fit og Predict model
'''
# fit model - stopper efter 200 ved lille udvikling og gemmer bedste resultat - vigtigt pga. ML modeller svinger meget
callback = EarlyStopping(min_delta = 0.001, patience = 200, mode = "min", restore_best_weights = True)
missing_value = -99

# masked funktion benyttet af Kuo i Deep Triangle
def masked_mse(missing_value):
    
    def custom_mse(y_true, y_pred):
    # assume 1st dimension is the number of samples
        keep= tfk.cast(tfk.not_equal(y_true, missing_value), tfk.floatx())
        mse = tfk.mean(tfk.square((y_pred-y_true)*keep), axis=2)

        return mse

    return custom_mse


model.compile(optimizer = Adam(learning_rate=0.0005, amsgrad = True),
            loss         = [masked_mse(missing_value),masked_mse(missing_value)],
            loss_weights = [0.5,0.5])

### beregner model
```
